```python
stress_map = {"Low": 0.33, "Medium": 0.55, "High": 0.8}
heart["Stress Level"] = heart["Stress Level"].map(stress_map)
Alchol_map = {"Low": 0.33, "Medium": 0.55, "High": 0.8}
heart["Alcohol Consumption"] = heart["Alcohol Consumption"].map(stress_map)
Excercise_map = {"Low": 0.33, "Medium": 0.55, "High": 0.8}
heart["Exercise Habits"] = heart["Exercise Habits"].map(stress_map)
Sugar_map = {"Low": 0.33, "Medium": 0.55, "High": 0.8}
heart["Sugar Consumption"] = heart["Sugar Consumption"].map(stress_map)
```

```python
heart['Gender']=heart['Gender'].map({'Male':1,"Female":0})
heart['Smoking']=heart['Smoking'].map({'Yes':1,"No":0})
heart['Family Heart Disease']=heart['Family Heart Disease'].map({'Yes':1,"No":0})
heart['Diabetes']=heart['Diabetes'].map({'Yes':1,"No":0})
heart['High Blood Pressure']=heart['High Blood Pressure'].map({'Yes':1,"No":0})
heart['Low HDL Cholesterol']=heart['Low HDL Cholesterol'].map({'Yes':1,"No":0})
heart['High LDL Cholesterol']=heart['High LDL Cholesterol'].map({'Yes':1,"No":0})
heart['Heart Disease Status']=heart['Heart Disease Status'].map({'Yes':1,"No":0})
```

```python
X_heart = heart.drop(columns=['Heart Disease Status'])
y_heart = heart['Heart Disease Status']
```

```python

# Cancer
X_cancer_train, X_cancer_test, y_cancer_train, y_cancer_test = train_test_split(
    X_cancer, y_cancer, test_size=0.2, random_state=42, stratify=y_cancer
)

# Diabetes
X_diabetes_train, X_diabetes_test, y_diabetes_train, y_diabetes_test = train_test_split(
    X_diabetes, y_diabetes, test_size=0.2, random_state=42, stratify=y_diabetes
)

# # Heart Disease
X_heart_train, X_heart_test, y_heart_train, y_heart_test = train_test_split(
    X_heart, y_heart, test_size=0.2, random_state=42, stratify=y_heart)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report, accuracy_score
```

```python
svm_cancer = make_pipeline(StandardScaler(), SVC(kernel='rbf', C=1))
svm_cancer.fit(X_cancer_train, y_cancer_train)

y_pred_cancer = svm_cancer.predict(X_cancer_test)
print("Cancer Dataset")
print(classification_report(y_cancer_test, y_pred_cancer))
```

```python
svm_diabetes = make_pipeline(StandardScaler(), SVC(kernel='rbf', C=1))
svm_diabetes.fit(X_diabetes_train, y_diabetes_train)

y_pred_diabetes = svm_diabetes.predict(X_diabetes_test)
print("Diabetes Dataset")
print(classification_report(y_diabetes_test, y_pred_diabetes))
```

```python
svm_heart = make_pipeline(StandardScaler(), SVC(kernel='rbf', C=1))
svm_heart.fit(X_heart_train, y_heart_train)
# self.feature_order[name] = X_heart.columns.tolist()
y_pred_heart = svm_heart.predict(X_heart_test)
print("Heart Diseases")
print(classification_report(y_heart_test, y_pred_heart))
```
